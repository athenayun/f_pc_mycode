### Setting database table

```python
class m01_all(peewee.Model):
    UniqueSequence = peewee.CharField()
    Gantry = peewee.CharField()
    Lane = peewee.CharField()
    Date_Time = peewee.CharField()
    CarType = peewee.CharField()
    Direction = peewee.CharField()
    
    
    
    class Meta:
        database = db
```
