```python
def calculation_for_group(series, total_num, group_col):
    new_list = []
    place_list = []
    if len(series) == 0 :
        if len(group_col) == 1 :
            for i in range(0, 5):
                per_var = '0.0%'
                place_name = 'nan'
                new_list.append(per_var)
                place_list.append(place_name)
        else :
            for i in range(0, 10):
                per_var = '0.0%'
                place_name = 'nan'
                new_list.append(per_var)
                place_list.append(place_name)
    elif len(series) == 5 :
        for i in range(0, 5):
            per_var = '{0}%'.format(round(float(series[i])/float(total_num)*100, 1))
            new_list.append(per_var)        
    else :
        for i in range(0, 10):
            per_var = '{0}%'.format(round(float(series[i])/float(total_num)*100, 1))
            new_list.append(per_var)
    # check if it's an empty dataframe
    tmp_df = pd.DataFrame(series).reset_index().drop([0], axis=1)   
    if tmp_df.empty :
        tmp_df = pd.DataFrame()
        tmp_df['place'] = place_list
        if len(group_col) == 2 :
            tmp_df['place_2'] = place_list
    tmp_df['percentage'] = new_list
    
    return tmp_df
```

```python
def trip_sorting(df, directions, calculation_for_group):
    weekday = [21, 22, 23, 24, 25]
    weekend = [26, 27]
    weeks = [weekday, weekend]
    direction_dict = {}
    direction_df_in = []
    direction_df_out = []
    direction_df_total = []
    grouping_col_list = [['in_location'], ['out_location'], ['in_location', 'out_location']]
    
    for dire in directions :
        week_dict = {}
        week_df_in = []
        week_df_out = []
        week_df_total = []
        
        
        for idx, week_type in enumerate(weeks) :
            groupby_dict = {}
            for group_col in grouping_col_list :
                groupby_dict['num_{0}'.format(group_col)] = df[(df.STOP == 'P') & (df.DIRECTION_ID == dire)&(df.ENTRY_TIME.dt.day.isin(week_type))].groupby(group_col).size().sort_values(0, ascending=False).sum()
                if len(group_col) != 2 :
                    groupby_dict['series_{0}'.format(group_col)] = df[(df.STOP == 'P') & (df.DIRECTION_ID == dire)&(df.ENTRY_TIME.dt.day.isin(week_type))].groupby(group_col).size().sort_values(0, ascending=False)[:5]
                else :
                    groupby_dict['series_{0}'.format(group_col)] = df[(df.STOP == 'P') & (df.DIRECTION_ID == dire)&(df.ENTRY_TIME.dt.day.isin(week_type))].groupby(group_col).size().sort_values(0, ascending=False)[:10]
                # distincting week type : weekday or weekend
                week_dict["tmp_{1}_{0}_df".format(idx, group_col)] = calculation_for_group(groupby_dict['series_{0}'.format(group_col)], groupby_dict['num_{0}'.format(group_col)], group_col)
                # concat two df(weekday and weekend) into one
                if len(group_col) != 2 :
                    if "in" in group_col[0] :
                        week_df_in.append(week_dict["tmp_{1}_{0}_df".format(idx, group_col)])
                    elif "out" in group_col[0] :
                        week_df_out.append(week_dict["tmp_{1}_{0}_df".format(idx, group_col)])
                elif len(group_col) == 2 :
                    week_df_total.append(week_dict["tmp_{1}_{0}_df".format(idx, group_col)])
                   
        direction_dict['{0}_bigger_in_df'.format(dire)] = pd.concat(week_df_in, axis=1)
        
        direction_dict['{0}_bigger_out_df'.format(dire)] = pd.concat(week_df_out, axis=1)
        direction_dict['{0}_bigger_total_df'.format(dire)] = pd.concat(week_df_total, axis=1)
        
        direction_df_in.append(direction_dict['{0}_bigger_in_df'.format(dire)])
        direction_df_out.append(direction_dict['{0}_bigger_out_df'.format(dire)])
        direction_df_total.append(direction_dict['{0}_bigger_total_df'.format(dire)])
    iterables = [directions, ['weekdays', 'weekend'], ['place', 'percentage']]
    col = pd.MultiIndex.from_product(iterables)    
    
    in_df = pd.concat(direction_df_in, axis=1)
    in_df.columns = col
    out_df = pd.concat(direction_df_out, axis=1)
    out_df.columns = col
    
    iterables = [directions, ['weekdays', 'weekend'], ['place_start', 'place_end', 'percentage']]
    col_t = pd.MultiIndex.from_product(iterables) 
    
    total_df = pd.concat(direction_df_total, axis=1)
    total_df.columns = col_t
    
    return in_df, out_df, total_df
```

```python
# using loop to do all the SA
def all_SA_trip(define_gantry_location_in, define_gantry_location_out, calculation_for_group, trip_sorting):
    SA_name = ["Taian", "Zhongli", "Xihu", "Xilou", "Nantou", "QingShui", "Huko", "Guanxi"]
    for sa in SA_name :
        if sa == "Zhongli":
            df_N = pd.read_csv('D://code//data//data_SA_north//{0}_N.csv'.format(sa), low_memory=False)
            df_S = pd.read_csv('D://code//data//data_SA_north//{0}_S.csv'.format(sa), low_memory=False)
            df = pd.concat([df_N, df_S])
        else :
            df = pd.read_csv('D://code//data//data_SA_north//{0}.csv'.format(sa), low_memory=False)
        
        df['ENTRY_TIME'] = pd.to_datetime(df['ENTRY_TIME'],infer_datetime_format=True)
        df['EXIT_TIME'] = pd.to_datetime(df['EXIT_TIME'],infer_datetime_format=True)
        df['O_TIME'] = pd.to_datetime(df['O_TIME'],infer_datetime_format=True)
        df['D_TIME'] = pd.to_datetime(df['D_TIME'],infer_datetime_format=True)
        
        df['in_location'] = df['O_CITY'].apply(define_gantry_location_in)
        df['out_location'] = df['D_CITY'].apply(define_gantry_location_out)
        
        if (sa == "Xilou") or (sa == "Guanxi") :
            directions = ["N"]
        else :
            directions = ['N', 'S']
        try :
            in_df, out_df, total_df = trip_sorting(df, directions, calculation_for_group)
        except :
            
        
        writer = pd.ExcelWriter('D://code//data//data_SA_north//analysis_result//{0}_OD.xlsx'.format(sa))
        in_df.to_excel(writer,'O_CITY')
        out_df.to_excel(writer,'D_CITY')
        total_df.to_excel(writer,'OD_CITY')
        writer.save()
        print('done')
```
