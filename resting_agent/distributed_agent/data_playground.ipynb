```python
def transform_time(x):
    time_str = str(x)
    if len(time_str)<6:
        time_str = "0{0}".format(time_str)
    splitat = 2
    h, m, s = time_str[:splitat], time_str[splitat:splitat+2], time_str[splitat+2:]
    time = "{0}:{1}:{2}".format(h, m, s)
    time_str = datetime.strptime(time, '%H:%M:%S')
    return time_str
```

```python
def split_date(x):
    date_str = str(x)
    splitat = 4
    splitto = 6
    Y, M, D = date_str[:splitat], date_str[splitat:splitto], date_str[splitto:]
    date = "{0}/{1}/{2}".format(Y, M, D)
    date_str = datetime.strptime(date, "%Y/%m/%d")
    return date_str
```
